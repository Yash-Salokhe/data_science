```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
airline = pd.read_excel('EastWestAirlines.xlsx', sheet_name='data')
airline 
```

```python
airline2 = airline.drop(['ID#'], axis=1)
airline2
```

```python
airline.info()
```

```python
#Normalizing the data 
from sklearn.preprocessing import StandardScaler
airline2_norm = StandardScaler().fit_transform(airline2)
airline2_norm
```

```python
#DBSCAN clustering
dbscan = DBSCAN(eps=1, min_samples=10)
dbscan.fit(airline2_norm)
```

```python
dbscan.labels_
```

```python
clusters = pd.DataFrame(dbscan.labels_, columns=['Clusters'])
clusters
```

```python
airline['Cluster_id']=clusters
airline
```

```python
airline.groupby('Cluster_id').agg(['mean']).reset_index()
```
